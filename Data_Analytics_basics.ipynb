```python
#countries.csv dataset
data=pd.read_csv("/content/countries.csv")
```

```python
data
```

```python
data.shape #check the number of rows a nd columns
```

```python
#compare the population growth in "USA" and "China".
US=data[data.country=="United States"]
```

```python
US
```

```python
China=data[data.country=="China"]
```

```python
China
```

```python
plt.plot(US.year,US.population)
plt.show()
```

```python
plt.plot(US.year,US.population/10**6)  #(In above plot the value 1e8 shows the numbers are big so divide the total population each number in series with 1 million or 10^6).
plt.plot(China.year,China.population/10**6)
plt.legend(["US","China"])
plt.xlabel("Year")
plt.ylabel("Population")
plt.show()
```

```python
#check the population growth of each country
plt.plot(US.year,US.population/US.population.iloc[0]*100)  #(In above plot the value 1e8 shows the numbers are big so divide the total population each number in series with 1 million or 10^6).
plt.plot(China.year,China.population/China.population.iloc[0]*100)
plt.legend(["US","China"])
plt.xlabel("Year")
plt.ylabel("Population growth(first year=100)")
plt.show()
```
